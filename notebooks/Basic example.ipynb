# Basic example

Import the required processes and libraries. `from pg_to_evalscript import convert_from_process_graph` loads the library that this notebook is part of.

```python
import os
import json
from pg_to_evalscript.conversion import convert_from_process_graph
import jsbeautifier
```

Define process graph

```python
def load_pg_json(path):
    path = os.path.abspath(path)
    with open(path, "r") as f:
        pg = json.load(f)
    return pg
```

Load the process graph, convert it to an evalscript and print the generated evalscript.

```python
pg = load_pg_json("../tests/process_graphs/gee_uc1_pol.json")

evalscripts = convert_from_process_graph(pg)
evalscript = evalscripts[0]["evalscript"]
evalscript_string = evalscript.write()
print(jsbeautifier.beautify(evalscript_string))
```
